### Classification and Testing
#### 1. Import training features and labels

```python
import doc2vec_series2csvArb
import pandas as pd
```

```python
train_df = pd.read_csv('bembeddings1024.csv')
train_df.head()
```

```python
dups = train_df.duplicated()
print(f"Number of duplicate rows = {dups.sum()}")
```

```python
train_df.drop_duplicates()
```

#### 2. Import testing data and remove redudant rows


```python
test_df = pd.read_csv("dev_labeled.tsv", sep="\t")
test_df.head()
```

```python
test_df['#3 country_label'].value_counts()
```

```python
test_df = test_df[(test_df['#3 country_label']== 'Egypt' ) | (test_df['#3 country_label']== 'Iraq') |
        (test_df['#3 country_label']== 'Saudi_Arabia') | (test_df['#3 country_label']== 'Algeria' ) | 
        (test_df['#3 country_label']== 'Oman' ) | (test_df['#3 country_label']== 'Libya' ) |
        (test_df['#3 country_label']== 'Syria' ) | (test_df['#3 country_label']== 'Morocco' ) |
       (test_df['#3 country_label']== 'United_Arab_Emirates' )]
test_df['#3 country_label'].value_counts()
```

```python
test_df.head()
```

#### 3. Preprocess text and extract features

```python
test = test_df['#2 tweet_content']
doc2vec_series2csvArb.extract_Arabic_Doc_Embeddings( docSeries = test , modelname = "1024PreProcessed", embeddingDim = 128, vocabMinFrequency = 1,  trainEpochs = 25)
```

#### 4. Import inferred data set based on training model 

```python
test_set = pd.read_csv('doc_embeddings.csv')
print(test_set.shape)
```

#### 5. Prepare test labels for evaluation 

```python
test_labels = test_df['#3 country_label'].astype('str')

test_labels= test_labels.replace({ 'Iraq' : 1 ,'Egypt' : 2,'Saudi_Arabia' : 3,'United_Arab_Emirates':4,'Algeria' : 5, 'Oman' : 6, 'Syria' : 7, 'Libya' : 8 , 'Morocco' : 9})
```

```python
labels = train_df["#3 country_label"]
train_df = train_df.drop('#3 country_label', axis = 1)
```

#### 7. Train and Evalaute first model: Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier( max_features="sqrt" , n_estimators =100)
clf = clf.fit(train_df,labels)
pred_test = clf.predict(test_set)
print(pred_test[:20])
print([test_labels][:20])
```

```python
from sklearn.metrics import accuracy_score, classification_report
acc_clf = accuracy_score(test_labels,pred_test)
print(f"Accuracy of clf is {acc_clf}")
print(classification_report(test_labels,pred_test))
```

#### 8. Train and Evaluate second model: Ensemble Support Vector Machine Bagging

```python
from sklearn import svm
from sklearn.ensemble import BaggingClassifier
clf1 = BaggingClassifier(base_estimator=svm.LinearSVC())
```

```python
clf1 = clf1.fit(train_df,labels)
pred_test1 = clf1.predict(test_set)
acc_clf1 = accuracy_score(test_labels,pred_test1)
print(f"Accuracy of clf1 is {acc_clf1}")
print(classification_report(test_labels,pred_test1))
```

#### 9. Train and Evaluate third model: Naive Bayes with normal(Gaussian) distribution assumed since continous

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.naive_bayes import GaussianNB
clf2 = GaussianNB()
clf2 = clf2.fit(train_df,labels)
pred_test2 = clf2.predict(test_set)
acc_clf2 = accuracy_score(test_labels,pred_test2)
print(f"Accuracy of clf2 is {acc_clf2}")
print(classification_report(test_labels,pred_test2))
```
